```python
# Importing the train_test_split library
from sklearn.model_selection import train_test_split
```

```python
# X variable containing all the columns of dataset except the last one (target column)
X=df.iloc[:,:-1]
```

```python
# Y variable containing the data of target variable
y=df['target']
```

```python
# train_test_split with test set size = 20%
X_train, X_test, y_train, y_test=train_test_split(X,y, test_size=0.20, random_state=45)
```

```python
# Model Building
from sklearn.tree import DecisionTreeClassifier
```

```python
d_tree=DecisionTreeClassifier()
```

```python
d_tree.fit(X_train, y_train)
```

```python
pred=d_tree.predict(X_test)
```

```python
#Importing accuracy metrics library and confusion matrix lib
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
#Model getting an accuracy of 81%
accuracy_score(pred, y_test)*100
```

```python
#Plotting confusion matrix
from sklearn import metrics
confusion_matrix=metrics.confusion_matrix(pred, y_test)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [False, True])

cm_display.plot()
plt.show()
```
